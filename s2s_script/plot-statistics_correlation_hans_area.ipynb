```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import xarray as xr
```

## 45 time values

```python
data45 = xr.open_dataarray('/home/esauvat/Documents/NORCE/heavy_files/s2s-hans_area-0.5-correlation_statistics-45_times.nc')
```

```python
data45
```

```python
norm45 = data45.sel(statistics='norm')
mean45 = data45.sel(statistics='mean')
std45 = data45.sel(statistics='std')
norm45stand = data45.sel(statistics='norm_stand')
```

```python
def plot45(data, stat) :
    ltimes = list(data['time'].values)
    values = [list(data.sel(time=num).values) for num in ltimes]

    plt.figure(figsize=(12, 6))
    sns.boxplot(data=values)

    plt.xticks(ticks=range(len(ltimes)), labels=ltimes)
    plt.xlabel("Lead times")
    plt.ylabel("Correlation " + stat)
    
    title = "Correlation " + stat + " for 45 days lead time"
    path = '/home/esauvat/Documents/NORCE/unseen-storm-forecasts/plots/s2s-distrib-correlation/'+"_".join(stat.split())+'_45-hans_area.png'

    plt.title(title)
    plt.savefig(path)

    plt.show()
```

```python
plot45(norm45, 'norm')
```

```python
plot45(mean45, 'mean')
```

```python
plot45(std45, 'standard deviation')
```

```python
plot45(norm45stand, 'standardized norm')
```
